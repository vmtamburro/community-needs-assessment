```python
import censusdata
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Analysis using the American Census Data (ACS) collected by the American Community Survey
# Ongoing survey by the U.S. Census Bureau that provides up-to-date info on demographics,
# economics, housing, and social characteristics of the U.S. population annually.

# Unlice Decennial census (counts population every 10 years), the ACS provides frequent, detailed,
# and consistent data on how people live, helping governments and orgs with program administration
# and evaluation.

# Compare with Washington Borough https://datacommons.org/place/geoId/3477270

sns.set(style="whitegrid")
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# ----------------------------
# 1. Define geographies
# ----------------------------
geographies = {
    'Washington Twp (Warren)': censusdata.censusgeo([('state', '34'), ('county', '041'), ('county subdivision', '77300')]),
    'Washington Borough': censusdata.censusgeo([('state', '34'), ('county', '041'), ('county subdivision', '77270')]),
    'ZIP 07882': censusdata.censusgeo([('zip code tabulation area', '07882')])
}
```

```python
# ----------------------------
# 2. Download variables
# ----------------------------

# 2a. Basic demographics: population, sex
demo_vars = [
    'B01001_001E', # Total population
    'B01001_002E', # Male total
    'B01001_026E'  # Female total
]

demo_data = {}
for name, geo in geographies.items():
    demo_data[name] = censusdata.download('acs5', 2023, geo, demo_vars)

demo_df = pd.concat(demo_data)
demo_df.index = geographies.keys()
demo_df.columns = ['Total Pop', 'Male Pop', 'Female Pop']
print(demo_df)
```

```python
# 2b. Median income & poverty
income_pov_vars = ['B19013_001E', 'B17001_002E']  # median HH income, below poverty
income_pov_data = {}
for name, geo in geographies.items():
    income_pov_data[name] = censusdata.download('acs5', 2023, geo, income_pov_vars)

income_pov_df = pd.concat(income_pov_data)
income_pov_df.index = geographies.keys()
income_pov_df.columns = ['Median HH Income', 'Below Poverty']

# Merge
df = pd.concat([demo_df, income_pov_df], axis=1)
df['% Poverty'] = df['Below Poverty'] / df['Total Pop'] * 100
df.to_csv('median_income_poverty')
print(df)
```

```python
# ----------------------------
# 3. Age Distribution (youth / working / seniors)
# ----------------------------
# ACS B01001 detailed: male 0-17: B01001_003E-B01001_006E, male 18-64: B01001_007E-B01001_025E, male 65+: B01001_020E etc.
age_vars = [f'B01001_{str(i).zfill(3)}E' for i in range(3,50)]  # columns B01001_003E to B01001_049E

age_data = {}
for name, geo in geographies.items():
    age_data[name] = censusdata.download('acs5', 2023, geo, age_vars)

age_summary = {}
for name, row in age_data.items():
    # youth 0-17
    youth_cols = list(range(3,7)) + list(range(27,31))  # male/female youth cols
    youth = row.iloc[:, [i-3 for i in youth_cols]].sum(axis=1)[0]
    # working 18-64
    working_cols = list(range(7,20)) + list(range(31,44))
    working = row.iloc[:, [i-3 for i in working_cols]].sum(axis=1)[0]
    # seniors 65+
    senior_cols = list(range(20,27)) + list(range(44,50))
    seniors = row.iloc[:, [i-3 for i in senior_cols]].sum(axis=1)[0]
    
    age_summary[name] = {'Youth': youth, 'Working Age': working, 'Seniors': seniors}

age_df = pd.DataFrame.from_dict(age_summary, orient='index')
age_df.to_csv('age_distribution')
print(age_df)
```

```python
# ----------------------------
# 4. Education (B15003)
# ----------------------------
edu_vars = [f'B15003_{str(i).zfill(3)}E' for i in range(1,26)]
edu_data = {}
for name, geo in geographies.items():
    edu_data[name] = censusdata.download('acs5', 2023, geo, edu_vars)

edu_summary = {}
for name, row in edu_data.items():
    hs_or_less = row.iloc[:,0:9].sum(axis=1)[0]
    bachelors_up = row.iloc[:,21:25].sum(axis=1)[0]
    edu_summary[name] = {'HS or less': hs_or_less, 'Bachelors+': bachelors_up}

edu_df = pd.DataFrame.from_dict(edu_summary, orient='index')
edu_df.to_csv('education')
print(edu_df)
```

```python
# ----------------------------
# 5. Employment by Industry (C24050)
# ----------------------------
# Example: high-level categories
# C24050_002E - C24050_029E cover different industry sectors
emp_vars = [f'C24050_{str(i).zfill(3)}E' for i in range(2,30)]
emp_data = {}
for name, geo in geographies.items():
    emp_data[name] = censusdata.download('acs5', 2023, geo, emp_vars)

# Convert to % of total employed
emp_df = pd.DataFrame({name: row.sum(axis=1)[0] for name, row in emp_data.items()}, index=['Total Employed']).T
emp_df.to_csv('employment_summary')
print(emp_df)
```

```python
# ----------------------------
# 6. Foreign-born & Language (B05002, B16001)
# ----------------------------
# Foreign-born: B05002_002E = native, B05002_003E = foreign born
fb_vars = ['B05002_002E','B05002_003E']
fb_data = {}
for name, geo in geographies.items():
    fb_data[name] = censusdata.download('acs5', 2023, geo, fb_vars)

fb_summary = {}
for name, row in fb_data.items():
    native = row.iloc[0,0]
    foreign = row.iloc[0,1]
    fb_summary[name] = {'Native': native, 'Foreign-born': foreign}

fb_df = pd.DataFrame.from_dict(fb_summary, orient='index')
fb_df['% Foreign-born'] = fb_df['Foreign-born'] / (fb_df['Native'] + fb_df['Foreign-born']) * 100
fb_df.to_csv('foreign_born')
print(fb_df)
```

```python
# ----------------------------
# 7. Housing Affordability (B25070 for rent as % income, B25003 tenure)
# ----------------------------
# Example: get % households paying >30% income on rent
rent_vars = [f'B25070_{str(i).zfill(3)}E' for i in range(1,12)]  # B25070_001E to B25070_011E
rent_data = {}
for name, geo in geographies.items():
    rent_data[name] = censusdata.download('acs5', 2023, geo, rent_vars)

rent_summary = {}
for name, row in rent_data.items():
    total = row.iloc[0,0]
    # columns for rent >30%: B25070_006E to B25070_011E
    high_rent = row.iloc[0,5:11].sum()
    rent_summary[name] = {'% Rent >30% Income': high_rent / total * 100}

rent_df = pd.DataFrame.from_dict(rent_summary, orient='index')
rent_df.to_csv('housing_affordability')
print(rent_df)
```

```python
# ----------------------------
# 9. Seniors 65+ Living Alone (ACS B09021)
# ----------------------------
senior_vars = [
    'B09021_001E',  # Total 65+
    'B09021_002E',  # Male 65+ living alone
    'B09021_003E'   # Female 65+ living alone
]

senior_data = {}
for name, geo in geographies.items():
    df = censusdata.download('acs5', 2023, geo, senior_vars)
    total_65 = df.iloc[0,0]
    male_alone = df.iloc[0,1]
    female_alone = df.iloc[0,2]
    total_alone = male_alone + female_alone
    pct_alone = total_alone / total_65 * 100 if total_65 > 0 else 0
    senior_data[name] = {
        'Total 65+': total_65,
        'Living Alone (Male)': male_alone,
        'Living Alone (Female)': female_alone,
        'Total Living Alone': total_alone,
        '% Living Alone': pct_alone
    }
senior_df = pd.DataFrame.from_dict(senior_data, orient='index')
print(senior_df)
```

```python
# ----------------------------
# 10. Single-parent Households (ACS B11005)
# ----------------------------
sp_vars = [
    'B11005_002E',  # Married-couple with children
    'B11005_003E',  # Male householder, no spouse, with children
    'B11005_005E'   # Female householder, no spouse, with children
]

sp_data = {}
for name, geo in geographies.items():
    df = censusdata.download('acs5', 2023, geo, sp_vars)
    married = df.iloc[0,0]
    single_male = df.iloc[0,1]
    single_female = df.iloc[0,2]
    single_parent = single_male + single_female
    total_with_kids = married + single_parent
    pct_single = single_parent / total_with_kids * 100 if total_with_kids > 0 else 0
    sp_data[name] = {
        'Married HH w/ kids': married,
        'Single Father HH': single_male,
        'Single Mother HH': single_female,
        'Total Single-Parent HH': single_parent,
        '% Single-Parent HH': pct_single
    }

sp_df = pd.DataFrame.from_dict(sp_data, orient='index')
print(sp_df)
```

```python
# ----------------------------
# Disability prevalence (C18108)
# ----------------------------
# Variables:
# C18108_001E = Total civilian noninstitutionalized population
# C18108_002E = With a disability
# ----------------------------

disability_vars = ['C18108_001E', 'C18108_002E']

disability_data = {}
for name, geo in geographies.items():
    disability_data[name] = censusdata.download('acs5', 2023, geo, disability_vars)

# Summarize
disability_summary = {}
for name, df_dis in disability_data.items():
    total = df_dis.iloc[0,0]
    with_disability = df_dis.iloc[0,1]
    pct = (with_disability / total * 100) if total > 0 else 0
    disability_summary[name] = {
        'Total Population': total,
        'With Disability': with_disability,
        'Disability Rate (%)': pct
    }

disability_df = pd.DataFrame.from_dict(disability_summary, orient='index')
print(disability_df)
```

```python
# ----------------------------
# Veteran Population (Counts)
# ----------------------------
# ACS Table B21001:
# B21001_001E = Total civilian population 18+
# B21001_002E = Veterans
# B21001_003E = Nonveterans
# ----------------------------

veteran_vars = ['B21001_001E','B21001_002E','B21001_003E']

veteran_data = {}
for name, geo in geographies.items():
    veteran_data[name] = censusdata.download('acs5', 2023, geo, veteran_vars)

# Summarize counts
veteran_counts = {}
for name, df_vet in veteran_data.items():
    total_18plus = int(df_vet.iloc[0,0])
    veterans = int(df_vet.iloc[0,1])
    nonveterans = int(df_vet.iloc[0,2])
    
    veteran_counts[name] = {
        'Total Adults 18+': total_18plus,
        'Veterans': veterans,
        'Nonveterans': nonveterans
    }

import pandas as pd
veteran_df = pd.DataFrame.from_dict(veteran_counts, orient='index')
print(veteran_df)
```

```python
# ----------------------------
# Uninsured Population (ACS B27010)
# ----------------------------
uninsured_vars = ['B27010_001E', 'B27010_002E', 'B27010_003E']

uninsured_data = {}
for name, geo in geographies.items():
    uninsured_data[name] = censusdata.download('acs5', 2023, geo, uninsured_vars)

# Summarize
uninsured_summary = {}
for name, df_un in uninsured_data.items():
    total_pop = df_un.iloc[0,0]
    insured = df_un.iloc[0,1]
    uninsured = df_un.iloc[0,2]
    pct_uninsured = (uninsured / total_pop * 100) if total_pop > 0 else 0
    uninsured_summary[name] = {
        'Total Population': total_pop,
        'Insured': insured,
        'Uninsured': uninsured,
        '% Uninsured': pct_uninsured
    }

import pandas as pd
uninsured_df = pd.DataFrame.from_dict(uninsured_summary, orient='index')
print(uninsured_df)
```
